### Matplotlib demo

```python
# esto es clave para que el grafico sea representado en el notebook
%matplotlib notebook

import matplotlib.pyplot as plt
import math

x = [i * 0.1 for i in range(0, 100)]

y1 = list(map(math.sin, x))
y2 = list(map(math.cos, x))

plt.plot(x,y1,'r')
plt.plot(x,y2,'b')
```
